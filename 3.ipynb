```python
import pandas as pd
import datasets

df = pd.read_csv("udemy_summer_honey_chips.csv")
df
```

```python
df = df.rename(columns={"レビュー":"text", "ラベル":"label"})
df = df.drop(columns="id")
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.4, random_state=11, stratify=df["label"])
df_test, df_val = train_test_split(df_test, test_size=0.5, random_state=11, stratify=df_test["label"])
```

```python
df_train.to_csv("./train.csv", index=False)
df_val.to_csv("./validation.csv", index=False)
df_test.to_csv("./test.csv", index=False)
```
