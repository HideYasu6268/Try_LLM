## テキスト分類

```python
from transformers import pipeline
```

```python
clf = pipeline("text-classification")
# clf = pipeline(model = "ProsusAI/finbert")
```

```python
# text = ["this stake is too salty", "this is nice"]
text = ["しょっぱい","これはおいしい！"]
output = clf(text)
output
```

## 質疑応答

```python
reader = pipeline("question-answering")
```

```python
context = "I am a professional baseball player"
question = "How much do I want?"

output = reader(question=question, context=context)
output
```

## 要約

```python
summarizer = pipeline("summarization")
```

```python
text = "Visited a burger joint renowned for its delicious pickles, and it truly lived up to the hype. The crunch and tanginess of the pickles perfectly complemented the juicy, well-seasoned patty, elevating the entire burger experience to a new level. The buns were soft yet sturdy enough to hold everything together, ensuring a mess-free enjoyment. Definitely a must-try for both burger enthusiasts and pickle lovers alike. A delightful culinary discovery."
```

```python
output = summarizer(text, min_length = 20, max_length = 40)
output
```
